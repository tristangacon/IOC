```python
fig, ax1 = plt.subplots(figsize=(12, 7))


ax1.plot(df_france['year'], df_france['France_Expenditure_Sports'], 
         color='black', linewidth=2, label='Sports Budget (Million )')
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Sports Budget (Million )', color='black', fontsize=12)
ax1.tick_params(axis='y', labelcolor='black')


ax1.set_xticks(df_france['year'].unique())


ax2 = ax1.twinx()
ax2.plot(df_france['year'], df_france['France_Medal_Count'], 
         color='blue', linewidth=2, label='Medal Count', marker='o')
ax2.plot(df_france['year'], df_france['France_Gold_Medal'], 
         color='gold', linewidth=2, label='Gold Medal Count', marker='o')
ax2.set_ylabel('Medal Count', fontsize=12, color='black')   
ax2.tick_params(axis='y', labelcolor='black')               


ax1.axvline(x=2024, color='gray', linestyle='--', linewidth=2)


plt.title('France Expenditure on Recreation and Sports', fontsize=16)


lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
legend = ax2.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left',
                    facecolor='white', edgecolor='black', frameon=True)


for text in legend.get_texts():
    text.set_color("black")


ax1.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 7))


ax1.plot(df_japan['year'], df_japan['Japan_Sports_Budget'], 
         color='black', linewidth=2, label='Sports Budget (Billion )')
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Sports Budget (Billion )', color='black', fontsize=12)
ax1.tick_params(axis='y', labelcolor='black')


ax1.set_xticks(df_japan['year'].unique())


ax2 = ax1.twinx()
ax2.plot(df_japan['year'], df_japan['Japan_Medal_Count'], 
         color='blue', linewidth=2, label='Medal Count', marker='o')
ax2.plot(df_japan['year'], df_japan['Japan_Gold_Medal'], 
         color='gold', linewidth=2, label='Gold Medal Count', marker='o')
ax2.set_ylabel('Medal Count', fontsize=12, color='black')   
ax2.tick_params(axis='y', labelcolor='black')              


ax1.axvline(x=2021, color='gray', linestyle='--', linewidth=2)


plt.title('Japan Annual Sports Budget', fontsize=16)


lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
legend = ax2.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')


ax1.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
df_retail = pd.read_excel(r'C:\Users\trist\OneDrive\Documents\College\Baruch\CIS9655\ProjectIdeas\DatasetsCollected.xlsx', sheet_name='Retail')
df_retail.head(7)
```

```python
df_retail['year'] = df_retail['year'].astype('int64')
```

```python
plt.figure(figsize=(10, 6))


plt.plot(df_retail['year'], df_retail['UK'], color='blue', label='UK')
plt.plot(df_retail['year'], df_retail['Japan'], color='red', label='Japan')
plt.plot(df_retail['year'], df_retail['Brasil'], color='green', label='Brasil')


uk_2012 = df_retail[df_retail['year'] == 2012]
plt.scatter(uk_2012['year'], uk_2012['UK'], color='blue', edgecolor='black', s=100, zorder=5)
japan_2021 = df_retail[df_retail['year'] == 2021]
plt.scatter(japan_2021['year'], japan_2021['Japan'], color='red', edgecolor='black', s=100, zorder=5)
brasil_2016 = df_retail[df_retail['year'] == 2016]
plt.scatter(brasil_2016['year'], brasil_2016['Brasil'], color='green', edgecolor='black', s=100, zorder=5)


plt.title('YoY Retail Sales', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Yearly % Change', fontsize=12)


plt.xticks(df_retail['year'].unique())

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
